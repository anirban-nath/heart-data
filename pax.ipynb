```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('heart.csv')
dataset.head(10)
```

```python
X= dataset.iloc[:,0:13].values
y= dataset.iloc[:,13].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 22)
X.shape
```

```python
from keras.models import Sequential
from keras.layers import regularizers, Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(6, kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01), activation = 'relu', input_dim = 13))

model.add(Dropout(0.5))

model.add(Dense(5, kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.001), activation = 'relu'))

model.add(Dropout(0.2))

model.add(Dense(3, kernel_initializer='he_normal', activation = 'relu'))

model.add(Dropout(0.1))

model.add(Dense(1, kernel_initializer='he_normal', activation='sigmoid'))
```

```python
#Compiling the neural network
model.compile(optimizer ='adam', loss='binary_crossentropy', metrics =['accuracy'])
```

```python
#Fitting the data to the training dataset
history = model.fit(X,y, validation_split=0.1, batch_size=8, epochs=500)

# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
